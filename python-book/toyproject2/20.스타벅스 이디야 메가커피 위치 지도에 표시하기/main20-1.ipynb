```python
#데이터 읽기
import pandas as pd

file_path = "07_24_05_P.xlsx"
df = pd.read_excel(file_path,sheet_name=None)
```

```python
#두개의 시트를 합쳐 하나의 데이터프레임으로 합치기
df = pd.concat(df,ignore_index=True)
df
```

```python
#column 이름 확인
for col in df.columns: 
    print(col) 
```

```python
#사업장명 데이터만 출력
print(df['사업장명'])
```

```python
#스타벅스, 이디야, 메가커피 이름이 포함된 매장만 다시 데이터프레임 생성 후 몇개인지 확인
스타벅스_df = df[df['사업장명'].str.contains('스타벅스')]
print("스타벅스 매장수:",len(스타벅스_df))

이디야_df = df[df['사업장명'].str.contains('이디야')]
print("이디야 매장수:",len(이디야_df))

메가커피_df = df[df['사업장명'].str.contains('메가엠지씨|메가커피')]
print("메가커피 매장수:",len(메가커피_df))
```

```python
#주소를 -> 좌표로 
# http://api.vworld.kr 사이트의 (Geocoder API 2.0 레퍼런스) API사용

import requests

def request_geo(road,road_type,api_key):
    url = 'http://api.vworld.kr/req/address?'
    params = 'service=address&request=getcoord&version=2.0&crs=epsg:4326&refine=true&simple=false&format=json&type='
    address = '&address='
    keys = '&key='

    if road_type == "road": #도로명주소
        page = requests.get(url+params+road_type+address+road+keys+api_key)
    elif road_type == "parcel": #지번주소
        page = requests.get(url+params+road_type+address+road+keys+api_key)
    
    json_data = page.json()
    if json_data['response']['status'] == 'OK':
        x = json_data['response']['result']['point']['x']
        y = json_data['response']['result']['point']['y']
        return x,y
    else:
        x = 0
        y = 0
        return x,y

MY_API_KEY = 'B3297F1A-FC77-37B4-BA2C-7C7B109E776C'

x,y = request_geo(스타벅스_df['소재지전체주소'].iloc[0],"parcel",MY_API_KEY)
print(x,y)
```

```python
# 매장(스타벅스or이디야or메가), 매장이름, x, y  의 순으로 엑셀파일에 저장

스타벅스주소_list = 스타벅스_df['소재지전체주소'].to_list()
스타벅스_매장이름_list = 스타벅스_df['사업장명'].to_list()

이디야주소_list = 이디야_df['소재지전체주소'].to_list()
이디야_매장이름_list = 이디야_df['사업장명'].to_list()

메가커피주소_list = 메가커피_df['소재지전체주소'].to_list()
메가커피_매장이름_list = 메가커피_df['사업장명'].to_list()

매장_list = []
매장이름_list = []
좌표x_list = []
좌표y_list = []
for i,스타벅스_주소 in enumerate(스타벅스주소_list) :
    매장_list.append("스타벅스")
    매장이름_list.append(스타벅스_매장이름_list[i])
    try:
        x,y = request_geo(스타벅스_주소,"parcel",MY_API_KEY)
        좌표x_list.append(x)
        좌표y_list.append(y)
        print(스타벅스_매장이름_list[i],x,y)
    except:
        좌표x_list.append(0)
        좌표y_list.append(0)
        print("주소변환 에러")

for i,이디야_주소 in enumerate(이디야주소_list) :
    매장_list.append("이디야")
    매장이름_list.append(이디야_매장이름_list[i])
    try:
        x,y = request_geo(이디야_주소,"parcel",MY_API_KEY)
        좌표x_list.append(x)
        좌표y_list.append(y)
        print(이디야_매장이름_list[i],x,y)
    except:
        좌표x_list.append(0)
        좌표y_list.append(0)
        print("주소변환 에러")

for i,메가커피_주소 in enumerate(메가커피주소_list) :
    매장_list.append("메가커피")
    매장이름_list.append(메가커피_매장이름_list[i])
    try:
        x,y = request_geo(메가커피_주소,"parcel",MY_API_KEY)
        좌표x_list.append(x)
        좌표y_list.append(y)
        print(메가커피_매장이름_list[i],x,y)
    except:
        좌표x_list.append(0)
        좌표y_list.append(0)
        print("주소변환 에러")

스타벅스_이디야_메가커피_df = pd.DataFrame()
스타벅스_이디야_메가커피_df["매장"] = 매장_list
스타벅스_이디야_메가커피_df["매장이름"] = 매장이름_list
스타벅스_이디야_메가커피_df["좌표_X"] = 좌표x_list
스타벅스_이디야_메가커피_df["좌표_Y"] = 좌표y_list
스타벅스_이디야_메가커피_df.to_excel("스타벅스_이디야_메가커피_위치.xlsx")
```
